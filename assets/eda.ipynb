## How many events and how many days in each event?

```python
# Count unique days for each event in dataset_df
events_days = dataset_df.groupby(['event', 'EventName'])['date'].nunique().reset_index()
events_days.columns = ['event', 'EventName', 'unique_days']
print("Number of days for each event:")
print(events_days)
```

## How many documents in each event?

```python
# Count the number of documents for each event
documents_per_event = dataset_df.groupby('event').size().reset_index(name='document_count')

print("Number of documents in each day:")
print(documents_per_event)
```

## How many queries in each event?

```python
# Count the number of queries for each event in queries_df
queries_per_event = queries_df.groupby('event_id').size().reset_index(name='query_count')
queries_per_event.columns = ['event_id', 'query_count']
print("Number of queries for each event:")
print(queries_per_event)
```

```python
# Rename 'event_id' in queries_per_event to match 'event' in events_days for merging
queries_per_event.rename(columns={'event_id': 'event'}, inplace=True)

# Merge the two DataFrames on the 'event' column
merged_chart = events_days.merge(queries_per_event, on='event', how='inner')
merged_chart = merged_chart.merge(documents_per_event, on='event', how='inner')

# Display the merged chart
print("Merged chart showing number of days, queries, and documents for each event:")
print(merged_chart)
```

## What's the time range for each events?

```python
# Convert unix_timestamp to datetime if not already done
dataset_df['datetime'] = pd.to_datetime(dataset_df['unix_timestamp'], unit='s')

# Find the time range for each event
time_range = dataset_df.groupby('event')['datetime'].agg(['min', 'max']).reset_index()
time_range.columns = ['event', 'start_time', 'end_time']
print("Time range of events:")
print(time_range)
```
